# Going through all of the products

```python
filename = "products.csv"
f = open(filename , "w")
headers = "brand, product_spec, price, shipping\n"
f.write(headers)
```

```python
for cont in containers:
    brand = cont.find("a",{"class":"item-brand"}).img["title"]
    spec= cont.img['title']
    price = cont.find("li",{"class":"price-current"}).strong.text+cont.find("li",{"class":"price-current"}).sup.text
    shipping = cont.findAll("li",{"class":"price-ship"})[0].text.strip()
    print("Brand: "+ brand)
    print("Prod.specification: "+ spec)
    print("Price: "+ price)
    print("shipping: "+ shipping)
    print('\n')
    f.write(brand + "," +spec.replace(",","|")+","+price+","+shipping+"\n")
f.close()
```
